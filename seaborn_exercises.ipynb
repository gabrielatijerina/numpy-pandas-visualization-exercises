```python
# Exercise 4 - Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.
```

```python
from env import host, user, password
```

```python
def get_db_url(user, host, password, db):
    url = (f'mysql+pymysql://{user}:{password}@{host}/{db}')
    return (url)
```

```python
chipotle_sql = '''
                SELECT *
                FROM orders;
                '''
```

```python
chipotle = pd.read_sql(chipotle_sql, get_db_url(user, host, password, 'chipotle'))
chipotle.head()
```

```python
#Convert item_price to float
chipotle['item_price'] = chipotle['item_price'].str.replace('$', '').astype(float)
```

```python
# Group by item name then get sum of quantity and price 
top_4 = chipotle.groupby('item_name')[['quantity', 'item_price']].sum().nlargest(4, 'quantity', keep = 'all')
top_4
```

```python
plt.bar(top_4.index, top_4.item_price, color = 'royalblue', ec = 'black')
plt.xticks(rotation = 45)
plt.xlabel('Menu Item')
plt.ylabel('Revenue Generated')
plt.title('The Big Four at Chipotle')
plt.show()
```
