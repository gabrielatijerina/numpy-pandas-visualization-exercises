```python
# Exercise 1

# Write a function called add_one that takes in a number and adds one to it.
    # Math notation for this would be y = x + 1
# Set your x variable to be range(1, 20)
# Using a loop or a list comprehension on your x variable, apply the add_one function to create our y variable.
# Then, use matplotlib to make plot a line that shows x and y.

def add_one(x):
    y = x + 1
    return y

x = range(0, 20)
y = [add_one(n) for n in x]

plt.plot(x,y)
plt.show()
```

```python
# Exercise 2
# Write a function called times_five that takes in a number and multiplies it by five.
# Set your x variable to be range(1, 20)
# Using a list comprehension on your x variable, apply the add_one function to create our y variable.
# Then, use matplotlib to make plot a line that shows x and y.

def times_five(x):
    y = x * 5
    return y

x = range(1, 20)
y = [times_five(n) for n in x]
plt.plot(x,y)
```

```python
# Exercise 3
# Write a function that does the following. Name the function three_and_five
    # For multiples of 3, multiply the number by 3
    # For multiples of 5, multiple the number by 5
    # For any other number, return the number itself
# Set your x variable to be range(1, 20)
# Using a list comprehension on your x variable, apply the three_and_five function to create our y variable.
# Then, use matplotlib to make plot a line that shows x and y.

def three_and_five(x):
    if x % 3 == 0:
        return x * 3
    elif x % 5 == 0:
        return x * 5
    else:
        return x
    
x = range(1, 20)
y = [three_and_five(n) for n in x]
plt.plot(x, y)
plt.show()
```

```python
# Exercise 4
# Write a function called square that takes in a number and squares it.
# Set your x variable to be range(-20, 20)
# Using a list comprehension on your x variable, apply the add_one function to create our y variable.
# Then, use matplotlib to make plot a line that shows x and y.

def square(x):
    y = x ** 2
    return y
x = range(-20, 20)
y = [square(n) for n in x]
plt.plot(x, y)
plt.show()
```
